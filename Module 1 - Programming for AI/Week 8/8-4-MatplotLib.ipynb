In the previous lesson we checked out various aspects of data processing using NumPy. In order to understand data, it is very useful to be able to visualize it while processing this.

This can be achieved in Python using the MatplotLib

In this lesson we explore various functionalities available to us through the MatplotLib package

Matplotlib.pyplot provides us with a set of functions that we can use for visualizing the patterns present in the data and is very useful when we explore any kind of data processing and analysis. 

There are a variety of functions that pyplot enables us to generate quick visualizations with ease. An example of using Pyplot for visualizing some data is given below:


```python
import numpy as np
import matplotlib.pyplot as plt
t = np.arange(0., 5., 0.2)

# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

In the above example we are using a graph with three curves comparing the differences between linear function $t$, $t^2$ and $t^3$.
As can be seen through this example, we are easily able to compare various trends using Pyplot.

Another option could be to visualise this as different sub-plots side-by-side. This can be obtained as follows


```python
import numpy as np
import matplotlib.pyplot as plt
t = np.arange(0., 5., 0.2)

fig,ax = plt.subplots(1,2)
ax[0].plot(t,t**2, label='t-squared')
ax[1].plot(t,t**3, label='t-cubed')
ax[0].set_title('Squared function')
ax[1].set_title('Cubed function')
ax[0].legend()
ax[1].legend()
```

Let us consider an example of having a sub-plot within a main plot. This can be achieved through the following way

```python
import numpy as np
import matplotlib.pyplot as plt
t = np.arange(0., 5., 0.2)

fig,ax = plt.subplots()
axins = ax.inset_axes([0.2,0.5,0.4,0.3] ) #relative left, bottom, width, height with respect to the main plot
ax.plot(t,t**2, label='t-squared')
axins.plot(t,t**3, label='t-cubed')

ax.set_title('Main plot')
axins.set_title('Inset plot')
ax.legend()
axins.legend()
```
