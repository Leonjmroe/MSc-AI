```python
import matplotlib.pyplot as plt

# Data
accuracies = [89.6, 93.8, 93.6, 95]
labels = ['Bayes', 'Logistic R', 'Combined', 'Combined Buffer']

# Create bar chart
plt.figure(figsize=(10, 6))
bars = plt.bar(labels, accuracies, color=['blue', 'green', 'red', 'purple'])  # Capture bars here

# Labels and title with bold font for axes
plt.xlabel('Model Transition', fontweight='bold')
plt.ylabel('% Change in Accuracy', fontweight='bold')
plt.title('Accuracy Change in Model Transitions')
plt.ylim(min(accuracies) - 1, max(accuracies) + 1)  # Adjust y-axis to include negative values

# Loop to overlay values inside bars
for bar, accuracy in zip(bars, accuracies):
    # Adjust text placement based on the value of accuracy (positive or negative)
    if accuracy >= 0:
        placement = bar.get_height() + 0.3  # Slightly above the bar for positive values
        va = 'bottom'
    else:
        placement = 0.3  # A fixed distance above the x-axis for negative values
        va = 'bottom'
    
    plt.text(bar.get_x() + bar.get_width() / 2, placement, f'{round(accuracy, 2)}%', ha='center', va=va, color='black', fontweight='bold')

# Show plot
plt.show()
```
